```python
import typing
from kdrag.all import *
from collections import defaultdict
class Clause:
    neg: frozenset[smt.ExprRef]
    pos: frozenset[smt.ExprRef]


def clause(xs):
    return sorted(xs, key=lambda l: l.get_id())




def ground_ordered_resolution(clauses):
    clauses = {max(l.get_id() for l in c) : c for c in clauses}
    
    todo = [clauses]
    passive = defaultdict(list)
    while todo:
        c = todo.pop()
        A = max(l.get_id() for l in c)
        for c1 in passive[A]:
            todo.append(frozenset(c + c1))
        












```
