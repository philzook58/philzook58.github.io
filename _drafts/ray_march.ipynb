```python
import matplotlib.pyplot
import jax
import jax.numpy as jnp


jnp.zeros(10) + 1

def sphere(p, r=1.0):
    return jnp.linalg.norm(p) - 1.0

@jax.jit
def raymarch(ray):
    p = jnp.zeros(3)
    for i in range(5):
        d = sphere(p)
        p = p + ray * d
    return jnp.abs(sphere(p)) < 0.01

```
