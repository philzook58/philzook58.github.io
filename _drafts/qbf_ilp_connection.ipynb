So say we've got a closed formula of QBL. We can put it in prenex form:
Qp₁...Qpₙ,φ(p₁,...pₙ)
It's true IFF there is binary branching tree, of depth n+1, which forks when Q is ∀, and doesn't when Q is ∃, where each non-root node is decorated with either a 1 or a 0, where the forks always give both a one and a zero, and where every branch is a satisfying assignment for the formula φ(p₁,...pₙ).
So like, ∀p, p∨¬p has its truth witnessed by the tree
           *
          ╱ ╲
   p     0   1
And ∀p∃q, p :left_right_arrow: q has its truth witnessed by
           *
          ╱ ╲
   p     0   1
         │   │
   q     0   1
So you just need to encode the approximate notion of one of these quantifier trees into the language of modal logic. The construction will be to build some modal sentences only collectively satisfiable in a Kripke frame that
has the right branching properties,
where each pₖ gets the right truth value by layer k, and keeps it in all later layers.
where φ(p₁,...pₙ) is true in every leaf
Strategy: if you need the tree to have depth n + 1, then for each k ≤ n, introduce an propositional variable dₖ which is supposed to be true at exactly the points in the kth layer of the tree (root is 0th). You can require this by writing:
□ᵏdₖ
□ᵏ¬dᵢ for i ≠ k
□ᵏdₖ → ◇dₖ₊₁
You want each branching layer to split, so if Qₖ is an ∀, you write
□ᵏdₖ → ◇pₖ₊₁ ∧ ◇¬pₖ₊₁
and otherwise
□ᵏdₖ → ◇pₖ₊₁ ∨ ◇¬pₖ₊₁
and you can ensure that truth values are passed on by adding
□ⁱ((pₖ → □pₖ) ∧ (¬pₖ → □¬pₖ))
for all i ≥ k.
Finally, you need □ⁿφ(p₁,...pₙ), to require that φ(p₁,...pₙ) is true at the leaves.
Jotted down quickly, probably there's an off-by-one or worse in there, but I think that's the idea. (edited)

OK, I've got something that I think works. But it involves subscripting variables with lists, and doing any amount of that would be probably look like a mess on slack. Here's the basic idea:
Work in the language ∧,∨,¬, so that you can drive negations down to literals - do that.
label each box and diamond with a unique index i.
Translate recursively, with a side-parameter for the list L of indicies
   traversed so far:
a. f(φ∧ψ,L) => f(φ,L) ∧ f(ψ,L)
b. f(φ∨ψ,L) => f(φ,L) ∨ f(ψ,L)
c. f(◇ᵢφ,L) => f(φ,L+i)
c. f(p,L) => pᴸ                   
c. f(~p,L) => ~pᴸ                   --in other words, decorate literals with the list
d. f(□ᵢφ,L) => boxTransform(φ,L+1) 
4. Existentially bind everything still unbound.
Step d in the recursive bit is the most complicated. boxTransform(φ,L+i) should mean: universally bind p decorated with L+i for each p in the original formula. Add a conditional whose antecedent is a disjunction of conjunctive clauses, one conjunction for each diamond index j, with conjunction j saying that each p decorated with L+i is equivalent to p decorated with L+j. The consequent existentially binds each p decorated with any L+i+j, for J a diamond index, and then continues recursively with the translation, with f(φ,L+i).
Here's a worked example. Equals means biconditional, the biconditional doesn't render well in my browser.
◇p ∧ ◇¬p ∧ □((s=p) ∧ ◇s)
=translation + simplification=>
∃p₁p₂[ p₁ ∧ ¬p₂ ∧ ∀p₃s₃[((p₃=p₁) ∧ (s₃=s₁)) ∨ ((p₃=p₂) ∧ (s₃=s₂)) → ∃s₃₊₄[(s₃=p₃) ∧ s₃₊₄]]]
A (valued) frame witnessing the satisfiability of the original modal formula on the LHS would need a point x where the formula is true and (because of the diamonds) two points accessible from x, x₁, where p is true and x₂, where p is false. Because of the box, x₁ needs to make s true,  x₂ needs to make s false, and both x₁ and x₂ need to access worlds x₁₊₁ and x₂₊₁ that make s true.
To show that the RHS is satisfiable, we need an assignment to propositional variables that witnesses its satisfiability. We can get this by taking the assignment  that makes a subscripted variable aᵢ true just in case a is true at xᵢ. This clearly takes care of the first two conjuncts. For the universal conjunct, because of the antecedent restricting the possible values of p₃, we just need to verify that this makes (a) ∃s₃₊₄, s₁=p₁ ∧ s₃₊₁ and (b) ∃s₃₊₁, s₂=p₂ ∧ s₃₊₁ true. We can do that by letting s₁₊₁ witness for s₃₊₄ in (a), and letting s₂₊₄ witness for s₃₊₄ in (b). Vice versa (constructing a frame from a satisfying assignment to the formula) is similar, introducing one point in the frame for each subscript made entirely of diamond-indicies, and reading accessibility relations off of the prefix-of relation on the subscripts.
