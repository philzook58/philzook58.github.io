Nested integegrators are like tries
telescopes




```python
import random
random.random()
N = 100


def integ(a,b):
    return lambda f: sum(f(a + random.random()*(b-a)) for _ in range(N)) * (b-a) / N
integ(0,1)(lambda x: x**2)
integ(0,1)(lambda x: integ(0,x)(lambda y: y))
```
