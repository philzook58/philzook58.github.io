```python
from kdrag.all import *
from kdrag.smt import *
import kdrag.theories.real as real
def dist(x,y):
    return real.abs(x - y)
q = Real("q")
S = smt.Const("S", smt.SetSort(RealSort()))
x,y,z = Reals("x y z")
f = Array("f", RealSort(), RealSort())
inside = kd.define("inside", [f, S], ForAll([x], S[x], S[f[x]]))
contracts = kd.define("contracts", [f, S], Exists([q], q > 0, q < 1, ForAll([x,y], S[x], S[y], dist(f[x], f[y]) <= q * dist(x,y))))

@kd.PTheorem(ForAll([f, S], contracts(f, S),
            Exists([z], f[z] == z)))
def fixed_point(l):
    f = l.fix()
    l.unfold(contracts)
    l.intros()

@kd.Theorem(contracts(smt.Lambda([x], x / 2)))
def half(l):
    l.unfold(contracts)
    l.unfold(real.abs)
    #l.exists(smt.RealVal(0.5))
    l.auto()

```
