```python
%%file /tmp/file.ssa
function w $add(w %a, w %b) {              # Define a function add
@start
	%c =w add %a, %b                   # Adds the 2 arguments
	ret %c                             # Return the result
}
export function w $main() {                # Main function
@start
	%r =w call $add(w 1, w 1)          # Call add(1, 1)
	call $printf(l $fmt, ..., w %r)    # Show the result
	ret 0
}
data $fmt = { b "One and one make %d!\n", b 0 }
```

```python

from typing import List, Tuple, Optional
from lark import Lark, Tree, Token

qbe_grammar = r"""
module: function*

function: "function" type FUNCNAME "(" [param_list] ")" "{" block* "}"

param_list: param ("," param)*
param: type TEMP

block: LABEL instr*

?instr: binop
     | jmp
     | jnz
     | ret

binop: TEMP "=" type BINOP TEMP "," TEMP
jmp: "jmp" LABEL
jnz: "jnz" TEMP "," LABEL "," LABEL
ret: "ret" TEMP

type: "w" | "l" | "s" | "d"  // TYPE

BINOP: "add"|"sub"|"mul"|"div"

// TYPE: /[wldp]/
FUNCNAME: /\$[A-Za-z_][A-Za-z0-9_]*/
TEMP: /%[A-Za-z_][A-Za-z0-9_]*/
LABEL: /@[A-Za-z_][A-Za-z0-9_]*/

%import common.WS
%ignore WS
%ignore /#.*/   // line comments
"""

parser = Lark(qbe_grammar, start="module", parser="lalr")

example = """
function w $add(w %a, w %b) {
@start
    %c =w add %a, %b
    ret %c
}
"""
from kdrag.all import *
parser.parse(example)

def instr(tree : Tree):
    match tree:
        case Tree("binop", [dest, type_, op, src1, src2]):
            sort = to_sort(type_)
            d,s1,s2 = [smt.Const(str(v), sort) for v in [dest, src1, src2]]
            table = {
                "add" : operator.add,
                "sub" : operator.sub,
                "mul" : operator.mul,
                "div" : operator.floordiv,
            }
            if op in table:
                return (d, table[op](s1,s2))
            

        
def to_sort(tree : Tree) -> smt.SortRef:
    match tree:
        case Tree("type", ["w"]):
            return smt.BitVecSort(32)
        case Tree("type", ["l"]):
            return smt.BitVecSort(64)
        case Tree("type", ["s"]):
            return smt.Float32Sort()
        case Tree("type", ["d"]):
            return smt.Float64Sort()
        case _:
            raise ValueError(f"Not a type: {tree}")

def block(tree : Tree):
    match tree:
        case Tree("block", [label, *instrs]):
            instrs = [instr(instr_tree) for instr_tree in instrs]
            # kd.define(label, instrs)
            return (label, instrs)
        case _:
            raise ValueError(f"Not a block: {tree}")

def function(tree : Tree):
    match tree:
        case Tree("function", [type_, name, params, *blocks]):
            env = {}
            blocks = [block(blk) for blk in blocks]
        case _:
            raise ValueError(f"Not a function: {tree}")
```
