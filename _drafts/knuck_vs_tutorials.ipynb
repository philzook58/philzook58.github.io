````python
import kdrag as kd
import kdrag.smt as smt
import kdrag.solvers as solvers
T = smt.DeclareSort('T')
B = smt.BoolSort()
x = smt.Const('x', B)
y = smt.Const('y', B >> B)
f = smt.Const("f", B >> (B >> B))
surjective = smt.Lambda([f], smt.ForAll([x], smt.Exists([y], y == f(x))))

thm = smt.Not(smt.Exists([f],smt.Select(surjective, f)))

s = solvers.EProverTHFSolver()
s.add(smt.Not(thm))
s.check()
````
