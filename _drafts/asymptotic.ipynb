```python
import sympy
from sympy import *

y = Function("y")
t = symbols("t")
eps = symbols("epsilon")
eq = y(t).diff(t,2) + y(t) + eps * y(t)**3

y0 = Function("y0")(t)
e0 = eq.subs({eps: 0, y(t) : y0})
sol0 = dsolve(e0)


res = Function("res")(t)
eq1 = eq.subs(y(t), sol0.rhs + eps * res).simplify().expand().collect(eps).coeff(eps)
sol1 = dsolve(eq1)


```

```python
y2 = Function("y2")(t)
e2 = eq.subs(y(t), sol0.rhs + eps * sol1.rhs + eps**2 * y2).expand().collect(eps).coeff(eps**2)
dsolve(e2, y2)
```
