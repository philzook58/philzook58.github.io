```python
from kdrag.all import *

def Choose(P):
    sorts = smt.domains(P)
    assert len(sorts) == 1, "Choose only supports predicates with one argument"
    return smt.Ext(P,smt.K(sorts[0], False))

s = smt.Solver()
x = smt.Int("x")
#s.add(x == Choose(smt.Lambda([x], x > 0)))
s.add(x == Choose(smt.SetAdd(smt.EmptySet(smt.IntSort()), smt.IntVal(9))))
s.add(x < 10)
s.check()
s.model()
```
