A fundamental confusion
`{a,b,c}` vs `a U b U c`

`f(a)` prolog vs `f(a)` atp
`f(a) = g(b)` prolog is syntactic, `f(a) = g(b)` atp is semantic

It's like when I counted 0,1,2,3,4 as a kid. That is a syntactic enumeration. The semmantic enumeration is 1,2,3,4,5

```python
@dataclass
class Universe:
    memo : dict[frozenset[Id], Id] = field(default_factory=dict)
    nodes : list[frozenset[Id]] = field(default_factory=list)

    def add(self, c : frozenset[Id]):
        if c not in self.memo:
            id = len(self.nodes)
            self.memo[c] = id
            self.nodes.append(c)
            return id
        else:
            return self.memo[c]
    def union(self, a : Id, b : Id):
        return self.add(self.nodes[a] | self.nodes[b])
    def intersection(self, a : Id, b : Id):
        return self.add(self.nodes[a] & self.nodes[b])
    def difference(self, a : Id, b : Id):
        return self.add(self.nodes[a] - self.nodes[b])
    def upair(self, a : Id, b : Id):
        return self.add(frozenset({a,b}))
    def sing(self, a : Id):
        return self.add(frozenset({a})) # or self.upair(a,a)
    def pair(self, a : Id, b : Id):
        return self.upair(self.upair(a,b), self.sing(a))
    def internalize(self):
        x = frozenset(self.memo.values())
        return self.add(x)

U = Universe()
emp = U.internalize()
one = U.internalize()
two = U.internalize()
three = U.internalize()

```
