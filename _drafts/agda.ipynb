```python
%%file /tmp/play2.agda
module play2 where

open import Data.Nat
open import Relation.Binary.PropositionalEquality using (_≡_; refl; cong)

myadd : ℕ → ℕ → ℕ
myadd zero m = ?
myadd (suc n) m = suc (myadd n m)
```

```python
! agda -i /tmp /tmp/play2.agda
```
