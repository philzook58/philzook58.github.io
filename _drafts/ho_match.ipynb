```python
%%file /tmp/test.smt2

(set-logic ALL)
(set-option :smt.mbqi false)
(set-option :ho_matching true)

(declare-fun f (Int) Int)
(declare-fun p (Int) Int)
(declare-fun Q (Int) Bool)

(assert
  (forall ((F (Array Int Int)))
    (! (Q (p (select F 0)))
       :pattern ((p (select F 0))))))

(assert (not (Q (p (f 0)))))

(check-sat)
```

```python
import z3
s = z3.Solver()
s.set(ho_matching=True)
s.add(z3.ForAll([z3.Array('F', z3.IntSort(), z3.IntSort())],
    z3.Implies(z3.Q(z3.Select(z3.Array('F', z3.IntSort(), z3.IntSort()), 0)),
        z3.Pattern(z3.Select(z3.Array('F', z3.IntSort(), z3.IntSort()), 0)))))
s.add(z3.Not(z3.Q(z3.Select(z3.Array('F', z3.IntSort(), z3.IntSort()), 0))))
print(s.check())
```
