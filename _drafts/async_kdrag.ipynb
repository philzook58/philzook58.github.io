```python
import subprocess
from kdrag.all import *
class Z3Solver(solvers.BaseSolver):
    #binary = z3.__file__ / "z3"
    binary = "z3"
    def __init__(self):
        self.proc = subprocess.Popen([self.binary, "-in" ], stdin=subprocess.PIPE, stdout=subprocess.PIPE)
        #sexpr = solver.to_sexpr()
        #proc.wait()
        super().__init__()
    def check(self):
        s = smt.Solver()
        for a in self.adds:
            s.add(a)
        self.proc.stdin.write(s.sexpr().encode())
        self.proc.stdin.write(b"(check-sat)")
        #print(s.sexpr())
        self.proc.stdin.close()
        #self.proc.communicate(stdin=s.sexp())
        retcode = self.proc.wait()

        if retcode == 0:
            out = self.proc.stdout.read().decode()
            if out == "unsat\n":
                return smt.unsat
            elif out == "sat\n":
                return smt.sat
            else:
                return smt.unknown        
        else:
            raise Exception("Z3 returned unknown result")



x, y = smt.Ints('x y')    
s = Z3Solver()
s.add(x + y > 5)
s.check()
```
