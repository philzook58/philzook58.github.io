````python
(declare-sort Term)
(declare-sort Var)
(declare-fun lam (String Term) Term)
(declare-fun app (Term Term) Term)
(declare-fun var (String) Term)

(declare-fun fresh (Term) String)

(declare-fun subst (Term String Term) Term)

(declare-fun free (Term String) Bool)

(assert (forall ((x String))
    (= (free (var x) x) true)))
(assert (forall ((x String) (y String))
    (=> (distinct x y) (= (free (var x) y) false))))
(assert (forall ((t Term) (x String))
    (= (free (lam x t) x) false)))
(assert (forall ((t Term) (x String) (y String))
    (=> (distinct x y) (= (free (lam x t) y) (free t y)))))
(assert (forall ((t1 Term) (t2 Term) (x String))
    (= (free (app t1 t2) x) (or (free t1 x) (free t2 x)))))

(assert (forall ((t Term) (x String) (s Term))
        (= (subst s x (var x) s))))
(assert (forall ((t Term) (x String) )))


````
