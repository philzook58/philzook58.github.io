```python
from kdrag.all import *
xor_decl = smt.Xor(smt.BoolVal(True), smt.BoolVal(False)).decl()
def is_xor(x):
    return x.decl() == xor_decl

def f2ify(x : smt.BoolRef):
    if smt.is_const(x):
        return x
    elif smt.is_not(x):
        return smt.Xor(smt.BoolVal(True), f2ify(x.arg(0)))
    elif smt.is_and(x):
        return smt.And([f2ify(arg) for arg in x.children()])
    elif smt.is_or(x):
        args = [smt.Xor(smt.BoolVal(True), y) for y in x.children()]
        return smt.Xor(smt.BoolVal(True), smt.And(args))
    elif is_xor(x):
        return smt.Xor(f2ify(x.arg(0)), f2ify(x.arg(1)))
    # implies
    else:
        raise ValueError("Unexpected formula", x)
    
def print_f2(x : smt.BoolRef):
    if smt.is_true(x):
        return "1"
    elif smt.is_false(x):
        return "0"
    elif smt.is_const(x):
        return str(x)
    elif is_xor(x):
        return f"({print_f2(x.arg(0))} + {print_f2(x.arg(1))})"
    elif smt.is_and(x):
        return f"({' * '.join([print_f2(arg) for arg in x.children()])})"

    else:
        raise ValueError("Unexpected formula", x)

x,y,z = smt.Bools("x y z")
print_f2(f2ify(x & y | z))


e = smt.And(x,y, smt.Or(z, smt.Not(x)))
kd.prove(e == f2ify(e))
```
