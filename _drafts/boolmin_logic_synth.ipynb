"
I started writing FBGB to speed up the groebner basis based minimization algorithm. For 5 variables you can compute maybe 10,000 groebner bases per second. For 8+ variable booleans it can take hundreds of seconds for just one boolean, even using SINGULAR or Polybori. Existing GB tools aren't designed for these dense systems we're processing. And there's special optimizations that work only for these huge, low variable count systems. The obvious optimization is treating each polynomial in the ideal as a truth table. You can cram 64 terms or members of a truth table into one u64, then everything is 64x faster. And if you treat a boolean polynomial as a truth table, addition is just bitwise XOR. Same for other ops. Another huge speedup came from linearization (substitute every monomial with a temporary variable implicitly) and and putting the system in RREF.  The whole gimmick of the groebner basis stuff in fbgb was to use a bunch of optimizations that only work for systems with small numbers of variables.

While writing the groebner basis stuff I started thinking about why the GB minimization algorithm stuff even works, and what are the lower / upper bounds on how compute a GB can be. That led me to write https://github.com/Colton1skees/anf-minimizer, which is basically trying to find the smallest bitwise OR of boolean polynomials that is equivalent to some input boolean. I then realized that GBs and boolean polynomials can become way small if you just negate some variables. a|b|c|d|e has like 32 terms in ANF, but if you negate the input and output yielding ~(a&b&c&d&~e). For some boolean minimization problems that matteo and I were stumped on, I was able to get optimal or near optimal results just by negating the variables and running this adhoc algorithm for minimizing polynomials in ANF.

Finding some combination of variable negations that minimizes the size of a boolean polynomial is 50% of the NPN canonicalization problem. so I figured I might aswell write an NPN library too. I could write a lot about this but the short version is that I do a brute force search through all 2^n combinations of negations to find the smallest anf poly, then an n! search through all variable orderings to find a canonical variable ordering. Heuristics everywhere to make this practical

That's pretty much the gist. I realized after the fact that the negation idea is identical to the linear sifting idea from BDDs
