```python
%%file /tmp/example.py
import sys
from rpython.rlib.jit import JitDriver

INC = "INC"
JMP = "JMP"

jitdriver = JitDriver(greens=['pc', 'prog'], reds=["acc"])

def run_loop(prog):
    acc = 0
    pc = 0
    while pc < len(prog):
        jitdriver.jit_merge_point(pc=pc,
                                  acc=acc,
                                  prog=prog)
        if acc % 1000000 == 0:
            print acc
        instr = prog[pc]
        if instr == INC:
            acc += 1
            pc += 1
        elif instr == JMP:
            pc = 0
            jitdriver.can_enter_jit(pc=pc, acc=acc, prog=prog)
        else:
            raise Exception("unknown instruction", instr)

def target(*args):
    return main

def main(args):
    # actually used  by rpython
    if len(args) != 2:
        print("usage: %s <progfile>" % args[0])
        return 1
    with open(args[1]) as f:
        prog = f.read().splitlines()
    run_loop(prog)
    return 0

if __name__ == "__main__":
    main(sys.argv)


```
