```python
from dataclasses import dataclass, field

type Id = tuple[float, int]
@dataclass
class MinUF:
    parents : list[Id] = field(default_factor=list)
    def makeset(self):
        id = (float("inf"), len(self.parents))
        self.parents.append(id)
        return id
    def find(self, x):
        m, xid = x
        while True:
            my, yid = self.parents[xid]
            m = min(my,m)
            if xid != yid:
                return (m, yid)
            xid = yid
    def union(self, x, y):
        mx, xid = self.find(x)
        my, yid = self.find(y)
        if xid == yid:
            return None
        else:
            if mx < my:
                self.parents[yid] = (mx, xid)
            else:
                self.parents[xid] = (my, yid)
```
