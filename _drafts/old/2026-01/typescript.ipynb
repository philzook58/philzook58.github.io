```typescript
import { init, sat } from "npm:z3-solver";

const { Context } = await init();
const { Solver, Int } = new Context('main');

const x = Int.const('x');
const y = Int.const('y');

const solver = new Solver();
solver.add(x.add(2).le(y.sub(10))); // x + 2 <= y - 10

if (await solver.check() !== sat) {
  throw new Error("couldn't find a solution")
}
const model = solver.model();
console.log(`x=${model.get(x)}, y=${model.get(y)}`);
```

```typescript
const ctx = new Context('main');
let s = new ctx.Solver();
s.add(ctx.Int.const('x').gt(ctx.Int.val(10)));
await s.check()
```

```typescript
const ctx = new Context('main');
//Int.const("x")
let z = ctx.Int.const("x")
let x = ctx.Real.fresh()
x.name()
x.toString()
x.le(1).sexpr()
x.eq(2).sexpr()
x.eqIdentity(x)
x.eqIdentity(z)

type Proof = {
    thm : ctx.Bool
}

async function prove(thm : ctx.Bool, by = []) : Proof {
    if(!ctx.isBool(thm)){
        throw new Error("Theorem must be a boolean expression");
    }
    let solver = new ctx.Solver();
    for (let lemma of by) {
        solver.add(lemma);
    }
    solver.add(thm.not());
    let result = await solver.check();
    console.log(result)
    if (result === "unsat") {
        return { thm };
    } else {
        throw new Error("Could not prove theorem");
    }
}
await prove(ctx.Bool.val(true));

```
