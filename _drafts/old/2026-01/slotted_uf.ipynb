```python
from dataclasses import dataclass, field

@dataclass
class UF():
    parents : list[int] = field(default_factory=list)
    def makeset(self):
        n = len(self.parents)
        self.parents.append(n)
        return n
    def find(self, a):
        while self.parents[a] != a:
            a = self.parents[a]
        return a
    def union(self, a, b):
        a = self.find(a)
        b = self.find(b)
        if a != b:
            self.parents[a] = b
        return b
    def is_eq(self,x,y):
        return self.find(x) == self.find(y)
    
uf = UF()
x,y,z = [uf.makeset() for _ in range(3)]
uf.union(x,y)
uf
```
