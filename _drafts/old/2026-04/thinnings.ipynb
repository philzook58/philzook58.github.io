"Now, for thinnings, we play the same game with respect to their monoidal (tensor) structure. A thinning value is a list of strips, where each strip is prime with respect to the monoidal product: if a strip can be decomposed as a tensoring, exactly one of the factors is the monoidal unit.

So the key question is what are the strips?

a strip can be the identity on a list chunk
a strip can be the codeletion of a list chunk
a strip can be a stuck thinning computation, as long as its wee end is nonempty (otherwise, it's bound to be a bunch of codeletions)
a strip can be a composition of thinning values (lists of strips), but only if (i) the aft end is not itself a single strip which happens to be a composition (so reassociation is compulsory), and (ii) there is no way to split the fore and aft lists in the same place (so interchange in favour of tensor is compulsory).
That's to say tensor is easy (it's just appending lists of strips), but composition is hard. Here's how you do it. Imagine that on two rows of an abacus, some fucker has been gluing some of the beads together. Your job is to split both rows into distinct regions of matching and minimal length. Whenever you have an overbite, you pull on the lower row; whenever you have an underbite, you pull on the upper row; anytime you match lengths, you've finished a region and are back in sync.

After that, you still have to reassociate the compositions and do any identity/codeletion arithmetic which presents itself (identity;identity = identity and codelete;identity = codelete).

I think this does the trick. I also like it because it's biased in favour of tensor and maximises detection of independence.

The args, *arg, and **kwargs models


This is a thinning model. We can lift known argument functions into ones that take more arguments but ignore them

````python
def weaken(f,thin):
    def res(*args):
        return f(*[a for b,a in zip(thin, args) if b])
    return res


def foo(x):
    return x+1

weaken(foo, [False, True, False])(1,65,100)
assert weaken(foo, [True, False, False])(1,65,100) == 2

weaken(foo, [True, True, False])(1,65,100)  # error
````
