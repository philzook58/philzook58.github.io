```rust
use std::collections::HashMap;
type Id = usize;
#[derive(Debug, PartialEq, Eq, Hash)]
enum Node {
    App(Id, Id),
    Lit(String)
}
#[derive(Debug)]
struct EGraph {
    memo : HashMap<Node, Id>,
    nodes: Vec<Node>,
}
```
